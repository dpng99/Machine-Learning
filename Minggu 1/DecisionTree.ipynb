```python
import pandas as pd
import numpy as np

rumah = pd.read_csv('test.csv')
kereta = pd.read_csv('train.csv')
data_rumah = pd.read_csv('melb_data.csv')
```

```python
y = data_rumah.Price
fitur_rumah = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude','Longtitude']
X = data_rumah[fitur_rumah]
X.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
train_X, val_X, train_y, val_y = train_test_split(X,y,random_state=0)
X.head()
```

```python
model_rumah = DecisionTreeRegressor()
model_rumah.fit(train_X, train_y)
```

```python
print("Prediksi 5 rumah")
print(val_X.head())
print("Prediksinya adalah")
print(model_rumah.predict(val_X.head()))
```
